```python
list_name = ['asep', 'amir', 'budi', 'cinta', 'edi', 'asep']

asep = []
for getIndex, getName in enumerate(list_name): # enumerate berfungsi untuk penomoran indeks
    if getName == 'asep':
        asep.append(getIndex)
# menampilkan indeks asep
print(asep)
```

```python
list_name = ['asep', 'amir', 'budi', 'cinta', 'edi', 'asep'] # daftar nama pada variabel list_name

asep = [] # list kosong buat nampung hasil
i = 0 # variabel 0 buat inisialisasi awal pada variabel i
for name in list_name: # untuk name dalam list name
  if name == 'asep': # jika name adalah asep
    asep.append(i) # maka si asep masuk ke list kosong dengan mengebalikan nilai dalam bentuk i (indeks)
  i += 1 # indeks saat ini +1 (kembali ke awal)

print(asep) # menampilkan indeks asep
```

list comprehension

```python
asep = [getIndex for getIndex, getName in enumerate(list_name) if getName == 'asep']
print(asep)
```

logic
- jika bilangan genap: tambah dengan 1
- jika bilangan ganjil: tambah dengan 2
- hasil penjumlahan masukan dalam list baru

```python
# looping secara umum
bilangan = [2, 4, 3, 6, 7, 9]

bilangan_baru = []
for i in bilangan:
  if i % 2 == 0:
      bilangan_baru.append(i + 1)
  else:
      bilangan_baru.append(i + 2)
print(bilangan_baru)
```

```python
# menggunakan list comprehension
hasil = [i + 1 if i % 2 == 0 else i + 2 for i in bilangan]
print(hasil)
```

```python
# menggunakan fungsi
def ganjil_genap(angka):
  bilangan_baru = []
  for i in angka:
      if i % 2 == 0:
          bilangan_baru.append(i+1)
      else:
          bilangan_baru.append(i+2)
  return bilangan_baru

bilangan = [2, 4, 3, 6, 7, 9]
print(ganjil_genap(bilangan))
```

```python
# menggunakan fungsi tanpa looping
def ganjil_genap(angka):
  if angka % 2 == 0:
    return(angka+1)
  else:
    return(angka+2)

bilangan = [2, 4, 3, 6, 7, 9]
bilangan_baru = list(map(ganjil_genap, bilangan))
print(bilangan_baru)
```

```python
def kuadrat(angka):
  return angka ** 2

angka_list = [1, 2, 3, 4, 5]
hasil = map(kuadrat, angka_list)

# Untuk melihat hasilnya, Anda perlu mengkonversinya ke list atau iterable lain
hasil_list = list(hasil)
print(hasil_list)  # Output: [1, 4, 9, 16, 25]
```

## Continue

```python
for i in range(10):
  if i % 2 == 0:
    continue
  else:
    print(i)
```

## Break

```python
for i in range(10):
  if i % 2 == 0:
    print(i)
  else:
    break
```
